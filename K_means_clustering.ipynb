```python
#K_means_clustering
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16,9)
plt.style.use('ggplot')
```

```python
#Importing the dataset
data = pd.read_csv("./mallCustomerData.txt", sep=",")
print(data.shape)
data.head(10)
```

```python
print(data["Gender"].value_counts())
```

```python
#Getting the values and plotting it
f1 = data["Annual Income (k$)"].values
f2 = data['Spending Score (1-100)'].values

for key in data.keys():
    print(key)

x = np.array(list(zip(f1,f2)))
# x = [[ 15  39] [ 15  81][ 16   6][ 16  77][ 17  40] ....]

#Dispersion graphic
plt.scatter(f1, f2, c="black", s=20)
```

```python
from sklearn.cluster import KMeans
#Number of clusters
kmeans = KMeans(n_clusters=3)
#Fitting the input data
kmeans = kmeans.fit(x)
#Getting the cluster labels
labels = kmeans.predict(x)
#Centroid values
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
from mpl_toolkits.mplot3d import Axes3D
''' fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x[:,0], x[:,1], x[:,1]) '''

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # Use fig.add_subplot instead of Axes3D
ax.scatter(x[:, 0], x[:, 1], x[:, 1])  # Assuming you have two features (0, 1)

plt.show()  # Display the plot

```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # Use fig.add_subplot instead of Axes3D
# Scatter plot for data points in clusters (color='y' for yellow)
ax.scatter(x[:, 0], x[:, 1], x[:, 1], c='y', label='Data Points')
# Scatter plot for centroids (marker='*' and color='#050505')
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 1], marker='*', c='red', s=1000, label='Centroids')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('Cluster Visualization')
ax.legend()

plt.show()  # Display the plot
```
